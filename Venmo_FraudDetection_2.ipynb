```python
import pandas as pd

# Path to the CSV file with headers
file1_path = '/content/venmo_short.csv'


# Read the CSV file with headers to get the column names
headers_df = pd.read_csv(file1_path)



print(headers_df)
```

```python
headers_df.columns = headers_df.columns.str.replace('.', '')
transaction_counts = headers_df.groupby(['paymenttargetuserusername', 'paymentactorusername']).size().reset_index(name='transactioncount')
result_df = pd.merge(headers_df, transaction_counts, on=['paymenttargetuserusername', 'paymentactorusername'], how='left')
print(result_df)


new_column_name = 'transactioncount'


def check_words(row):
    if row[new_column_name]>1:
        return 1
    else:
        return 0

result_df['is_fradulent'] = result_df.apply(check_words, axis=1)


print(result_df)
result_df.to_csv('output_file.csv', index=False)
```

```python
cursor.query("""
  CREATE OR REPLACE FUNCTION DrugDetection_XGBoost FROM
  ( SELECT * FROM drug_usage )
  TYPE XGBoost
  PREDICT 'is_fradulent'
  TIME_LIMIT 180
  METRIC 'accuracy'
  TASK 'classification';
""").df()
```

```python
cursor.query("SELECT DrugDetection_XGBoost(*) FROM drug_usage;").df()
```

```python
cursor.query("""
  SELECT is_fradulent, transactioncount_predictions FROM drug_usage
  JOIN LATERAL DrugDetection_XGBoost(*) AS Predicted(transactioncount_predictions);
""").df()
```
